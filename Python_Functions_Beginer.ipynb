````python
# 9. Write a function `is_even(num)` that returns:
#     - `"Even"` if the number is even.
#     - `"Odd"` otherwise
def is_even(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

result = is_even(int(input("Enter a Number: ")))
print(result)
````
