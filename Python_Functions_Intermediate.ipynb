```python
# 1. Write a function `factorial(num)` that returns the factorial of a number.

def factorial(num):
    fact = 1

    while num > 0:
        fact = fact * num
        num -= 1

    return fact

print(factorial(int(input("Enter a Num: "))))
```

```python
# 2. Write a function `sum_of_digits(num)` that returns the sum of the digits.
def sum_of_digits (num):
  total = 0
  while num > 0:
    digit = num % 10
    total = total + digit
    num = num // 10
  return total
print (sum_of_digits (234))
```

```python
# 3. Write a function `count_digits(num)` that returns the number of digits.
def count_digits(num):
    count = 0
    while num > 0:
        count += 1
        num = num // 10
    return count

print(count_digits(5647))
```

```python
# 4. Write a function `reverse_number(num)` that returns the reversed number.
def reverse_number(num):
  rev = 0
  while num > 0 :
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10
  return rev
print (reverse_number(695478))
```

```python
# 5. Write a function `is_palindrome(num)` that returns:
#     - `"Palindrome"`
#     - `"Not Palindrome"`
def is_palindrome (num):
  original_num = num
  pal = 0
  while num > 0 :
    digit = num % 10
    pal = pal * 10 + digit
    num = num // 10
  if original_num == pal :
      return ("Palindrome")
  else :
      return ("Not Palindrome")
print (is_palindrome(125))
```
