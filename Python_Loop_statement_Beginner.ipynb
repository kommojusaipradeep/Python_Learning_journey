```python
# 1.Print numbers from 1 to 10.
for i in range (1,11):
  print (i)
```

```python
# 2.Print numbers from 10 to 1.
for i in range (10,0,-1):
  print (i)
```

```python
# 2.Print numbers from 10 to 1.
i = 10
while i>=1 :
  print (i)
  i -= 1
```

```python
# 3.Print all even numbers between 1 and 20.
for i in range (2,21,2):
  print (i)
```

```python
# 3.Print all even numbers between 1 and 20.
for i in range (1,21):
  if i % 2 == 0 :
    print (i)
```

```python
# 4.Print all odd numbers between 1 and 20.
for i in range (1,21):
  if i % 2 != 0 :
    print (i)
```

```python
# 4.Print all odd numbers between 1 and 20.
for i in range (1,21,2):
  print (i)
```

```python
# 5.Print the multiplication table of 7. ( Example - 7 x 1 = 7)
#Multiplication of numbers
a = int(input("Enter a number "))
for i in range (1,11):
  print ( a,"X", i , "=" , a * i )
```

```python
# 5.Print the multiplication table of 7. ( Example - 7 x 1 = 7)
#using f string
a = int(input("Enter a number "))
for i in range (1,11):
  print (f"{a} X {i} = {a * i}")
```

```python
# 6.Print each character of the string: ( name="Python" )
name = "Python"
for ch in name:
  print (ch)
```

```python
# 7.Find the sum of numbers from 1 to 100.
total = 0
for i in range (1,101) :
  total = total + i
print (total)

# If print(total) is inside the loop, it prints the running total every time.
# Placing it outside the loop prints only the final sum.
#print(total)
```

```python
# Write a program to count how many even numbers are there between 1 and 100. Don't sum them—just count them.
count = 0
for i in range (1,101):
  if i % 2 == 0 :
    count += 1
print (count)
```

```python
# 8.Print numbers from 1 to 20, but skip 10 using continue.
for i in range (1,21):
  if i == 10 :
    continue
  print (i)
```

```python
# 9.Print numbers from 1 to 20, but stop when the number becomes 12 using break.
for i in range (1,21):
  if i == 12 :
    break
  print (i)
```

```python
# 10.Print the square of numbers from 1 to 10.
for i in range (1,11):
  print (i**2)
```
