```python
# 1. Count how many vowels are present in: word="programming"
word = "programming"
count = 0

for char in word:
    if char in "aeiou":
        count += 1

print(count)
```

```python
# 2. Find the factorial of a number.
num = int(input("Enter a Number: "))
fact = 1

for i in range(num, 0, -1):
    fact *= i

print("Factorial =", fact)
```

```python
# 3. Reverse a string using a loop.
#     Example: Python
#     Output: nohtyP

word = "Python"
rev_word = ""
for ch in word:
  rev_word = ch + rev_word
print (rev_word)
```

```python
# 7. Check whether a number is Prime.
#   Example: 13
#   Output: Prime Number
num = int(input("Enter a Number: "))

if num <= 1:
    print("Not Prime")
else:
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime Number")
    else:
        print("Not Prime Number")
```

```python
# 8. Find the largest number in:
#   numbers= [12,45,2,89,34]
numbers = [12, 45, 2, 89, 34]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest number is:", largest)
```

```python
# 10. Find the sum of all even numbers from 1 to 100.
total = 0

for i in range(101):
    if i % 2 == 0:
        total += i

print(total)
```
