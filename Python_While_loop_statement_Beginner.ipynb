```python
# Python_While_loop_statement_Beginner
# 1. Print numbers from **1 to 10**.
# 2. Print numbers from **10 to 1**.
# 3. Print all **even numbers** from 1 to 20.
# 4. Print all **odd numbers** from 1 to 20.
# 5. Print the multiplication table of a given number.
# 6. Find the **sum of numbers** from 1 to 100.
# 7. Find the **factorial** of a number.
# 8. Count the **digits** in a number.
# 9. Reverse a number.
# 10. Find the **sum of digits** of a number.
```

```python
# 1. Print numbers from **1 to 10**.
i = 1
while i<=10 :
  print (i)
  i += 1
```

```python
# 2. Print numbers from **10 to 1**.
i = 10
while i >=1:
  print (i)
  i -= 1
```

```python
# 3. Print all **even numbers** from 1 to 20.
i = 1
while i <=20 :
  if i % 2 == 0 :
    print (i)
  i += 1
```

```python
# 4. Print all **odd numbers** from 1 to 20.
i = 1
while i <=20 :
  if i % 2 != 0 :
    print (i)
  i += 1
```

```python
# 5. Print the multiplication table of a given number.
num = int(input("Enter a Number: "))
count = 1

while count <= 10:
    print(num, "x", count, "=",(num*count))
    count += 1
```

```python
 #6. Find the **sum of numbers** from 1 to 100.
i = 1
total = 0

while i <= 100:
    total = total + i
    i += 1

print(total)
```

```python
# 7. Find the **factorial** of a number.
num = int(input("Enter a number: "))
fact = 1

while num >0 :
  fact = fact * num
  num -= 1
print (fact)
```

```python
# 8. Count the **digits** in a number.
num = int(input("Enter a number: "))
count = 0

while num > 0:
    count += 1
    num = num // 10

print("Number of digits:", count)
```

```python
# 9. Reverse a number.
num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reversed number:", reverse)
```

```python
# 10. Find the **sum of digits** of a number.

num = int(input("Enter a number: "))
total = 0

while num > 0:
    digit = num % 10
    total = total + digit
    num = num // 10

print("Sum of digits:", total)
```
