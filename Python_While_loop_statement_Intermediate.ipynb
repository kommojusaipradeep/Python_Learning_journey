5. Check whether a number is an Armstrong number (3-digit)

An Armstrong number is a number whose digits cubed and summed equal the original number.

Example:

Input: 153

Calculation:
1³ + 5³ + 3³ = 1 + 125 + 27 = 153

Output:
Armstrong Number

Hint: Extract each digit, cube it (digit ** 3), add it to a total, and compare the total with the original number.

```python
# 1. Check if a number is a palindrome
# A palindrome number reads the same forwards and backwards.
# Example:
# Input: 121 Output: Palindrome
# Input: 123 Output: Not Palindrome
# Hint: Reverse the number first, then compare it with the original.

num = int(input("Enter a number: "))
original = num
pal = 0

while num > 0:
    digit = num % 10
    pal = pal * 10 + digit
    num = num // 10

if pal == original:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# 2.Find the largest digit in a number
# Example:
# Input: 593742 Output: 9
# Hint: Extract one digit at a time using % 10 and keep track of the largest digit.

num = int(input("Enter a number: "))
lar = 0

while num > 0:
    digit = num % 10

    if digit > lar:
        lar = digit

    num = num // 10

print("Largest digit:", lar)
```

```python
# 3.Find the smallest digit in a number
# Example:
# Input: 593742 Output: 2
# Hint: Initialize the smallest digit correctly and compare each extracted digit.

num = int(input("Enter a number: "))
smallest = 9

while num > 0:
    digit = num % 10

    if digit < smallest :
        smallest = digit

    num = num // 10

print("Smallest digit:", smallest)
```

```python
# 4.Count how many even and odd digits are in a number
# Example:
# Input: 987654
# Output: Even digits: 3 Odd digits: 3
# Hint: Extract each digit and check digit % 2.

num = int(input("Enter a number: "))
even_num = 0
odd_num = 0

while num > 0:
    digit = num % 10

    if digit % 2 == 0:
        even_num += 1
    else:
        odd_num += 1

    num = num // 10

print(f"Count of even digits: {even_num}")
print(f"Count of odd digits: {odd_num}")
```

```python
# 5.Check whether a number is an Armstrong number (3-digit)
# An Armstrong number is a number whose digits cubed and summed equal the original number.
# Example:
# Input: 153
# Calculation: 1³ + 5³ + 3³ = 1 + 125 + 27 = 153
# Output: Armstrong Number
# Hint: Extract each digit, cube it (digit ** 3), add it to a total, and compare the total with the original number.

num = int(input("Enter a number: "))
total = 0

while num > 0:
    digit = num % 10
    total = total + (digit ** 3)
    num = num // 10

print(total)
```
