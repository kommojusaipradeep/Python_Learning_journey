```python
#Intermediate
# 1.Find the larger of two numbers.
# 2.Check whether a year is a leap year.
# 3.Check if a student has passed (marks ≥ 35).
# 4.Check if a character is a vowel or a consonant.
# 5.Check if a number is a multiple of both 3 and 5.
```

```python
#1.Find the larger of two numbers.
num1 = int(input("Enter First Number: "))
num2 = int(input("Enter Second Number: "))

if num1 > num2:
    print("Num1 is larger")
elif num1 == num2:
    print("Both are Equal")
else:
    print("Num2 is larger")
```

```python
# 2.Check whether a year is a leap year.
#If the year is divisible by 400, it is a leap year.
#If the year is divisible by 4 but not by 100, it is a leap year.
#Otherwise, it is not a leap year.
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("It is a Leap Year")
else:
    print("It is Not a Leap Year")
```

```python
# 3.Check if a student has passed (marks ≥ 35).
marks = int(input("Enter your marks "))
if marks >= 35 :
  print ("Pass ")
else :
  print ("Fail ")
```

```python
# 4.Check if a character is a vowel or a consonant.
#.lower() is a string method in Python that converts all uppercase letters to lowercase.
letter = input("Enter a letter: ").lower()

if letter in ('a', 'e', 'i', 'o', 'u'):
    print("It is a vowel")
else:
    print("It is a consonant")
```

```python
#5.Check if a number is a multiple of both 3 and 5.
number = int(input ("Enter a number "))
if number % 3 == 0 and number % 5 == 0:
    print("The number is a multiple of both 3 and 5")
else:
    print("The number is not a multiple of both 3 and 5")
```
