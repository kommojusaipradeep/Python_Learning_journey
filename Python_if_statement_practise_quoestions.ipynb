```python
#1.Assign grades based on marks:
#90–100 → A
#75–89 → B
#60–74 → C
#35–59 → D
#Below 35 → Fail

marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
elif marks >= 35:
    print("Grade D")
else:
    print("Fail")
```

```python
# 2.Build a simple calculator:
#Ask for two numbers.
#Ask for an operation (`+`, `-`, `*`, `/`).
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
oper = input("Enter an operator (+, -, *, /): ")

if oper == "+":
    print("Result:", num1 + num2)
elif oper == "-":
    print("Result:", num1 - num2)
elif oper == "*":
    print("Result:", num1 * num2)
elif oper == "/":
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Division by zero is not allowed.")
else:
    print("Please select a valid operator (+, -, *, /).")0
```

```python
# 3.Compare three numbers and print the largest.
num1 = int(input("Enter num1: "))
num2 = int(input("Enter num2: "))
num3 = int(input("Enter num3: "))

if num1 > num2:
    if num1 > num3:
        print("Num1 is the largest")
    elif num1 == num3:
        print("Num1 and Num3 are the largest")
    else:
        print("Num3 is the largest")

elif num2 > num1:
    if num2 > num3:
        print("Num2 is the largest")
    elif num2 == num3:
        print("Num2 and Num3 are the largest")
    else:
        print("Num3 is the largest")

else:  #num1==num2 since above two conditions are false then only possible way is both are equal
    if num1 > num3:
        print("Num1 and Num2 are the largest")
    elif num3 > num1:
        print("Num3 is the largest")
    else:
        print("All three numbers are equal")
```

```python
# 5.Determine the type of triangle
#Equilateral Triangle - All sides are Equal
#Isosceles Triangle - Any two sides are equal
#Scalene Triangle - All are different sides
side1 = int(input("Enter side1: "))
side2 = int(input("Enter side2: "))
side3 = int(input("Enter side3: "))

if side1 == side2 == side3:
    print("Equilateral Triangle")
elif side1 == side2 or side2 == side3 or side1 == side3:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle")
```

```python
# 7.Calculate an electricity bill:
#First 100 units: ₹0
#Next 100 units: ₹5/unit
#Above 200 units: ₹10/unit

units = int(input("Enter number of units consumed: "))

if units <= 100:
    print("Your electricity bill is ₹0")
elif units <= 200:
    print("Your electricity bill is ₹", (units - 100) * 5)
else:
    print("Your electricity bill is ₹", (units - 200) * 10 + (100 * 5))
```
